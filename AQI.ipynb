```python
import requests

api = "a8bc7fcc0d461539d83668ce9d7389ed"  # Replace with your actual OpenWeatherMap API key

# 1. Ask for the location name instead of latitude/longitude
location_name = input("Enter the city or location name (e.g., Hyderabad): ")

# 2. Fetch coordinates using the Geocoding API
geo_url = "http://api.openweathermap.org/geo/1.0/direct"
geo_params = {
    'q': location_name,
    'limit': 1,
    'appid': api
}
geo_response = requests.get(url=geo_url, params=geo_params).json()

# 3. Check if location exists and proceed
if len(geo_response) > 0:
    lat = geo_response[0]['lat']
    lon = geo_response[0]['lon']
    print(f"Location found: {geo_response[0]['name']} (Lat: {lat}, Lon: {lon})")

    # 4. Fetch the Air Pollution data
    url = "http://api.openweathermap.org/data/2.5/air_pollution"
    parameters = {
        'lat': lat,
        'lon': lon,
        'appid': api
    }
    response = requests.get(url=url, params=parameters)
    response_json = response.json()

    print("Air pollution data fetched successfully! You can now run the next cell.")
else:
    print("Sorry, location not found. Please check the spelling and try running this cell again.")
    # Creating an empty fallback so the next cell doesn't completely crash if you accidentally run it
    response_json = {'list': [{'components': {'co': 0, 'no2': 0, 'o3': 0, 'so2': 0, 'pm2_5': 0, 'pm10': 0, 'nh3': 0}}]}
```

```python
components = response_json['list'][0]['components']
co = (components['co'])/1000
aqi = []
```

```python
def aqi_calculator(conc,bphi,bplow,aqihi,aqilow):
    aqi = ((aqihi - aqilow)/(bphi - bplow)) * (conc - bplow) + aqilow
    return aqi
```

```python
aqi_no2 = 0
if 0 <= components['no2'] <= 40:
   aqi_no2 = aqi_calculator(components['no2'],40,0,50,0)
elif 41 <= components['no2'] <= 80:
   aqi_no2 = aqi_calculator(components['no2'],80,41,100,51)
elif 81 <= components['no2'] <=  180:
   aqi_no2 = aqi_calculator(components['no2'],180,81,200,101)
elif 181 <= components['no2'] <= 280:
   aqi_no2 = aqi_calculator(components['no2'],280,181,300,201)
elif 281 <= components['no2'] <= 400:
   aqi_no2 = aqi_calculator(components['no2'],400,281,400,301)
aqi.append(aqi_no2)
```
